# #Ques 1-<br>
How has honey production yield changed from 1998 to 2021 ?

```python
plt.figure(figsize=(12,8))
sns.lineplot(data=df,x="year",y="yieldpercol")
plt.title("Yield changed from 1998 to 2021")
plt.xlabel("Year")
plt.ylabel("Yieldpercol")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(data=df,x="year", y="totalprod")
plt.title("total production changed from 1998 to 2021")
plt.xlabel("Year")
plt.ylabel("Total production")
plt.show()
```

# Ques 2-<br>
Overtime, what are the major production trends across the states ?

```python
plt.figure(figsize=(24,8))
sns.barplot(x=df['State'],y=df['totalprod'],ci=None, hue=df['year'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(6,3))
g=sns.FacetGrid(df,col='State',col_wrap=4,size=3)
g=g.map(plt.plot,'year','totalprod')
plt.show()
```

# Observation

# Question 3
Does the data show any trends in terms of the number of honey producing colonies and yield per colony before 2006 , which was when concern over Colony Collapse Disorder spread nation wide ?

```python
df_before_2006=df[df['year']<2006]
```

```python
df_before_2006.head()
```

```python
df_before_2006.shape
```

```python
State_colonies=df_before_2006.groupby('State')['numcol'].sum()
State_yield=df_before_2006.groupby('State')['yieldpercol'].mean()
```

```python
State_colonies.head(10)
```

```python
State_yield.head(10)
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10,8))
ax[0].plot(State_colonies.index, State_colonies.values)
ax[0].set_title("Number of Honey Producing Colonies (pre-2006)")
ax[1].plot(State_yield.index, State_yield.values)
ax[1].set_title("Yield per colony (pre-2006)")
plt.xticks(rotation=90)
plt.show()
```

#observation<br>
State wise no. of honey producing colonies<br>
State wise yield pe colony
